```python
sheet_url = "https://docs.google.com/spreadsheets/d/1sqoMPCahkw59_kxv9FFkHGEP6lrEJKdL3drDEkEZZZA/export?format=csv&gid=161960758"
sheet_df = pd.read_csv(sheet_url)
```

```python
fps = 30
expected_dt = 1 / fps
tolerance = 0.002
```

```python
for csv_path in sheet_df["csv_path"]:
    csv_path = str(csv_path).strip().strip('"').strip("'")
    print(f"\n=== {csv_path} ===")

    df = pd.read_csv(csv_path)

    df["Timestamp"] = pd.to_datetime(df["Timestamp"], utc=True)
    df["dt"] = df["Timestamp"].diff().dt.total_seconds()

    df["dropped"] = df["dt"] > (expected_dt + tolerance)
    df["n_dropped_est"] = np.round(df["dt"] / expected_dt - 1).clip(lower=0)

    dropped_frames = df[df["dropped"]]

    print(
        f"dropped intervals: {len(dropped_frames)} | "
        f"est frames dropped: {int(dropped_frames['n_dropped_est'].sum())}"
    )

    if len(dropped_frames) > 0:
        print(dropped_frames[["Timestamp", "dt", "n_dropped_est"]])
    df["dropped"] = df["dt"] > (expected_dt + tolerance)
    df["n_dropped_est"] = np.round(df["dt"] / expected_dt - 1).clip(lower=0)
    jitter_events = df[df["dropped"]]
    drop_events = df[df["n_dropped_est"] >= 1]
    print(f"jitter intervals: {len(jitter_events)}")
    print(f"true drop events: {len(drop_events)}")
    print(f"est frames dropped total: {int(drop_events['n_dropped_est'].sum())}")
    print("Drop events at recorded frame indices:")
    print(drop_events.index.tolist())
    missing = []
    for i, k in zip(drop_events.index, drop_events["n_dropped_est"].astype(int)):
        missing.extend(range(i + 1, i + 1 + k))

    print("Estimated dropped frame indices:")
    print(missing)


```
